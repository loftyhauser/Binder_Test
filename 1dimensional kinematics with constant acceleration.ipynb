```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Solving the 1-D equations of motion
$$x = x_0 + v_0t + 1/2at^2$$
$$v = v_0 + at$$

```python
x0=0
v0=50
a=-25
tinitial=0
tfinal=3
deltat=0.01
```

```python
t = np.arange(tinitial,tfinal,deltat)
v = v0+a*t
x = x0+v0*t+1/2*a*t**2
```

```python
plt.subplot(2,1,1)
plt.plot(t, x)
plt.ylabel('position (m)')
plt.title('Particle')

plt.subplot(2,1,2)
plt.plot(t, v)
plt.ylabel('velocity (m/s)')
plt.xlabel('time (s)')
```
