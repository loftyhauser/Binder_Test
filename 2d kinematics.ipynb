```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Solving the 2-D equations of motion
$$y = y_0 + v_{0y}t + 1/2a_yt^2$$
$$x = x_0 + v_{0x}t + 1/2a_xt^2$$
$$v_y = v_{0y} + a_yt$$
$$v_x = v_{0x} + a_xt$$

```python
x0=0
y0=0
v0=20
angle=50
ax=0
ay=-9.81
```

```python
t = np.arange(0,3,0.01)
v0y = v0*np.sin(np.deg2rad(angle))
v0x = v0*np.cos(np.deg2rad(angle))
vy = v0y+ay*t
vx = v0x+ax*t
x = x0+v0x*t+1/2*ax*t**2
y = y0+v0y*t+1/2*ay*t**2
```

```python
fix, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='x position (m)', ylabel='y position (m)', 
       title='Object position')
ax.grid()
```
